```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Task 2

**Q1**

```python
data = pd.read_csv('output/task2/q1.csv')

plt.bar(data['business_month'], data['count'], log=True)
plt.title('Trip counts for each business in each month')
plt.ylabel('Trip counts (log)')
plt.xlabel('Business-month')

plt.show()
```

**Q2**

```python
data = pd.read_csv('output/task2/q2.csv')

colors = ['red' if x<0 else 'blue' for x in data['profit']]
plt.bar(data['business_month'], abs(data['profit']), color=colors, log=True)
plt.title('Profits for each business in each month')
plt.ylabel('Profit (log)')
plt.xlabel('Business-month')
red = plt.Rectangle((0, 0), 1, 1, color='red', label='-ve Profit')
blue = plt.Rectangle((0, 0), 1, 1, color='blue', label='+ve Profit')
plt.legend(handles=[red, blue])

plt.show()
```

**Q3**

```python
data = pd.read_csv('output/task2/q3.csv')

plt.bar(data['business_month'], data['driver_earning'], log=True)
plt.title('Driver earnings for each business in each month')
plt.ylabel('Driver earning (log)')
plt.xlabel('Business-month')

plt.show()
```
